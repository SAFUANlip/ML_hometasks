## 1.1 Pure models.
First let's try to solve this case with good old Logistic Regression and simple (linear kernel) SVM classifier.

Train LR and SVM classifiers (choose params by hand, no CV or intensive grid search neeeded) and plot their decision regions. Calculate one preffered classification metric.

Describe results in one-two sentences.

_Tip:_ to plot classifiers decisions you colud use either sklearn examples ([this](https://scikit-learn.org/stable/auto_examples/neural_networks/plot_mlp_alpha.html#sphx-glr-auto-examples-neural-networks-plot-mlp-alpha-py) or any other) and mess with matplotlib yourself or great [mlxtend](https://github.com/rasbt/mlxtend) package (see their examples for details)

_Pro Tip:_ wirte function `plot_decisions` taking a dataset and an estimator and plotting the results cause you want to use it several times below

```python
def plot_decisions(X, y, clf):
    clf.fit(X,y)
    #plt.title(type(clf), f' accuracy: {accuracy_score(y, clf.predict(X))}')
    plt.title(f' accuracy: {accuracy_score(y, clf.predict(X))}, {type(clf)}')
    plot_decision_regions(X = X, y = y, clf = clf, legend = 'best')
    plt.show()
```

```python
lr = LogisticRegression(C = 0.4177, penalty = 'l1', solver='liblinear', tol=1e-3, max_iter=500) # add some params
svm = SVC(kernel='linear', C = 5, random_state=1) # here too

plot_decisions(moons_points, moons_labels, lr)
### YOUR CODE HERE
```

```python
plot_decisions(moons_points, moons_labels, svm)
```

## 1.2 Kernel tirck

Now use different kernels (`poly`, `rbf`, `sigmoid`) on SVC to get better results. Play `degree` parameter and others.

For each kernel estimate optimal params, plot decision regions, calculate metric you've chosen eariler.

Write couple of sentences on:

* What have happenned with classification quality?
* How did decision border changed for each kernel?
* What `degree` have you chosen and why?

```python
### YOUR CODE HERE


cv = StratifiedKFold(n_splits=10)
parameters = {'degree': np.arange(1,10), 'C': np.arange(1, 10), 'gamma': [0.001, 0.01, 1], "probability" : [True, False]}
for i in ['poly', 'rbf', 'sigmoid']:
    svm = SVC(kernel=i)
    grid = GridSearchCV(estimator=svm,
                 param_grid=parameters, cv = cv)
    grid=grid.fit(moons_points, moons_labels)
    grid.score(moons_points, moons_labels)
    print(grid.best_score_)
    print(grid.best_estimator_)
    print(grid.estimator.degree)
    plot_decisions(moons_points, moons_labels, grid.best_estimator_)
    #pd.DataFrame(grid.cv_results_)[['mean_test_score', 'std_test_score', 'params']].sort_values(by='mean_test_score', ascending = False)
    
```

## Answer. Best accuracy give RBF, best parameters C=9, degree=1, gamma=1, probability=True

## 1.3 Simpler solution (of a kind)
What is we could use Logisitc Regression to successfully solve this task?

Feature generation is a thing to help here. Different techniques of feature generation are used in real life, couple of them will be covered in additional lectures.

In particular case simple `PolynomialFeatures` ([link](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html)) are able to save the day.

Generate the set of new features, train LR on it, plot decision regions, calculate metric.

* Comare SVM's results with this solution (quality, borders type)
* What degree of PolynomialFeatures have you used? Compare with same SVM kernel parameter.

```python
### YOUR CODE HERE
for i in range(1,10):
    poly = PolynomialFeatures(degree = i)
    lr = LogisticRegression()
    pipe = Pipeline([('polynomial_features', poly), ('logistic_regression', lr)])
    pipe.fit(moons_points, moons_labels)
    pipe.score(moons_points, moons_labels)

    print(f"degre: {i}")
    plot_decisions(moons_points, moons_labels, clf = pipe)
    #plt.scatter(moons_points[:, 0], moons_points[:, 1], c = pipe.predict(moons_points)) 
    #plt.show()
```

## Compare with SVC kernel = poly, degree = 3

```python
svm = SVC(kernel='poly', degree=3, tol=0.0001, C=1.0, class_weight=None, random_state=None, max_iter=500, verbose=0) #standart parametrs from lr
plot_decisions(moons_points, moons_labels, clf=svm)
```

## 1.4 Harder problem

Let's make this task a bit more challenging via upgrading dataset:

```python
from sklearn.datasets import make_circles

circles_points, circles_labels = make_circles(n_samples=500, noise=0.06, random_state=42)

plt.figure(figsize=(5, 5))
plt.scatter(circles_points[:, 0], circles_points[:, 1], c=circles_labels)
```

```python
### YOUR CODE HERE
cv = StratifiedKFold(n_splits=10)
parameters = {'degree': np.arange(1,10), 'C': np.arange(1, 10), 'gamma': [0.001, 0.01, 1], "probability" : [True, False]}
for i in ['poly', 'rbf', 'sigmoid']:
    svm = SVC(kernel=i)
    grid = GridSearchCV(estimator=svm,
                 param_grid=parameters, cv = cv)
    grid=grid.fit(circles_points, circles_labels)
    grid.score(circles_points, circles_labels)
    print(grid.best_score_)
    print(grid.best_estimator_)
    print(grid.estimator.degree)
    plot_decisions(circles_points, circles_labels, grid.best_estimator_)
    #pd.DataFrame(grid.cv_results_)[['mean_test_score', 'std_test_score', 'params']].sort_values(by='mean_test_score', ascending = False)
    
```

## best it is accuracy 0.942 SVC(C=2, degree=6, gamma=1, kernel='poly', probability=True)

```python
parameters = {'penalty':['l2'], 'C':np.linspace(0.01, 100.0, 50), 
              "multi_class": ['auto', 'ovr', 'multinomial'], 
             } #"solver": ['lbfgs', 'liblinear', 'sag', 'saga'],
cv = StratifiedKFold(n_splits=5)
log_reg = LogisticRegression(max_iter= 2000)
for i in range(4,10):
    poly = PolynomialFeatures(degree = i)
    circles_points_poly = poly.fit_transform(circles_points)

    grid = GridSearchCV(estimator=log_reg,
                 param_grid=parameters, cv = cv)

    grid = grid.fit(circles_points_poly, circles_labels)
    grid.score(circles_points_poly, circles_labels)
    print(grid.best_score_)
    print(grid.best_estimator_)
    print(f"degree of poly: ", i)
    pipe = Pipeline([('polynomial_features', poly), ('logistic_regression', grid.best_estimator_)])
    plot_decisions(circles_points, circles_labels, pipe)
```

##  all models almost equal, so I can get one of them
